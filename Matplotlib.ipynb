```python
"""
Scipy:
We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:
1.fitting it to the periodic function
2.plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18
"""
```

```python
# Importing required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
MaxTemp = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
MinTemp = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])


months = np.arange(12)
plt.plot(months, MaxTemp, 'ro')
plt.plot(months, MinTemp, 'go')
plt.xlabel('Month')
plt.ylabel('Maximum & Minimum Temperature')
```

```python
from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,
                                      MaxTemp, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,
                                      MinTemp, [-40, 20, 0])
```

```python
# Creating a plot

days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(months, MaxTemp, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'g-')
plt.plot(months, MinTemp, 'go')
plt.plot(days, yearly_temps(days, *res_min), 'c-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

```python
# Creating a pie chart presenting the male/female proportion
s=round((titanic['sex'].value_counts())/len(titanic)*100,2)
pd.DataFrame(s)
```

```python
# Creating a plot

labels = ['male','female']
sizes = titanic.sex.value_counts()
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', colors = ['blue','green'])
plt.show()
```

```python
# Creating a scatterplot with the Fare paid and the Age, differ the plot color by gender

plt.figure()
category1 = titanic[titanic.sex=='male'].plot.scatter('age', 'fare', color='blue',label='male')
titanic[titanic.sex=='female'].plot.scatter('age', 'fare',color='green',label='female',ax=category1)
```
